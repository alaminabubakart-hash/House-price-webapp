```python
!pip install gradio
import gradio as gr
from sklearn.linear_model import LinearRegression
import pandas as pd

# Dataset
data = {
    "size_sqft":[500,800,1000,1200,1500,1800,2000,2500],
    "price":[50,80,100,130,160,190,210,260]
}

df = pd.DataFrame(data)

X = df[["size_sqft"]]
y = df["price"]

model = LinearRegression()
model.fit(X,y)

# Prediction function
def predict_price(size):
    prediction = model.predict([[size]])
    return f"Estimated price: {prediction[0]:.2f}"

app = gr.Interface(
    fn=predict_price,
    inputs=gr.Number(label="House Size (sqft)"),
    outputs=gr.Textbox(label="Predicted House Price"),
    title="🏠 House Price Predictor",
    description="Enter house size and get an estimated price using Machine Learning."
)

app.launch()
```
